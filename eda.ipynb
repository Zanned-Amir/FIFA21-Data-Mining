```python

class football:
    """Football is the class for exploratory data analysis and machine learning 
    on the football player dataset. This class has 8 important attributes which are as follows:
    
    multi_categorical_plot : plot multiple categorical data
    distplot_multi : plot the distribution of multiple data
    boxplot_multi : plot box plots for multiple data
    learner_selection : model selection for learning
    training_evaluate : evaluation of training
    """
    
    def __init__(self, data=None, cols=None, name='potential'):
        
        self.name = name # target
        self.data = data # feature
        self.cols = cols # feature columns name
        self.listof_model = {
            'LinearRegression': LinearRegression(), 
            'KNeighborsRegression': KNeighborsRegressor(),
            'RandomForestRegression': RandomForestRegressor(),
            'GradientBoostingRegression': GradientBoostingRegressor(),
            'XGBoostRegression': XGBRegressor(),
            'adaboost': AdaBoostRegressor()
        } # list of different learners
        
    # Read a CSV file
    def read(self, file):
        return pd.read_csv(file, sep=';', index_col='player_id')
    
    def multi_categorical_plot(self, data):
        """Plot a categorical feature
        
        data: float64 array  n_observation x n_feature
        """
        # Find a feature whose type is object
        string = []
        for i in data.columns:
            if data[i].dtypes == "object":
                string.append(i)
    
        fig = plt.figure(figsize=(20, 20))
        fig.subplots_adjust(wspace=0.4, hspace=0.3)
        for i in range(1, len(string) + 1):
            ax = fig.add_subplot(3, 1, i)
            sns.countplot(y=string[i-1], data=data, ax=ax)
            ax.set_title(f"{string[i-1]} countplot")
            
    def distplot_multi(self, data):
        """Plot multiple distplots"""
    
        from scipy.stats import norm
        plt.style.use('dark_background')
        cols = []
    
        # Feature whose type is int64 or float64
        for i in data.columns:
            if data[i].dtypes == "float64" or data[i].dtypes == 'int64':
                cols.append(i)
    
        gp = plt.figure(figsize=(20, 20))  
        gp.subplots_adjust(wspace=0.4, hspace=0.4)
        for i in range(1, len(cols) + 1):
            ax = gp.add_subplot(2, 3, i)
            sns.distplot(data[cols[i - 1]], fit=norm, kde=False, color='orangered', fit_kws={"color": "blue"}) 
            ax.set_title('{} '.format(cols[i - 1]))
            
    def boxplot_multi(self, data):
        """Plot multiple boxplots with a hue to represent categorical data."""
        
        plt.style.use('Solarize_Light2')
        plt.rcParams['text.color'] = 'black'
        cols = []
        for i in data.columns:
            if data[i].dtypes == "float64" or data[i].dtypes == 'int64':
                cols.append(i)
    
        gp = plt.figure(figsize=(20, 20))
        gp.subplots_adjust(wspace=0.4, hspace=0.4)
        for i in range(1, len(cols) + 1):
            ax = gp.add_subplot(2, 2, i)
            sns.boxplot(x=cols[i-1], data=data)
            ax.set_title('Boxplot for {}'.format(cols[i-1]))
            
    def correlation_plot(self, data, vrs='price'):
        """This function plots only one variable that is correlated to a target.
        
        data : m_observation x n_feature array
        vrs : target variable (n_observation, )
        cols : interested features
        """
        
        cols = []
        for i in data.columns:
            if data[i].dtypes == "float64" or data[i].dtypes == 'int64':
                cols.append(i)
                
        feat = list(set(cols) - set([vrs]))
    
        fig = plt.figure(figsize=(15, 10))
        fig.subplots_adjust(wspace=0.3, hspace=0.25)
        for i in range(1, len(feat) + 1):
        
            gp = data.groupby(feat[i-1]).agg('mean').reset_index()
        
            if len(feat) < 3:
                ax = fig.add_subplot(1, 3, i)
            else:
                n = len(feat) // 2 + 1
                ax = fig.add_subplot(2, n, i)
            
            ax.scatter(data[feat[i-1]], data[vrs], alpha=.25)
            ax.plot(gp[feat[i-1]], gp[vrs], 'r-', label='mean', linewidth=1.5)
            ax.set_xlabel(feat[i-1])
            ax.set_ylabel(vrs)
            ax.set_title('Plotting data {0} vs {1}'.format(vrs, feat[i-1]))
            ax.legend(loc='best')
            
    def split_data(self):
        """This function splits the data into training set and target set.
        
        data: feature matrix of dimensions n_observation x n_feature
        name: target (n_observation, )
        cols: interested features
        returns xtrain, xtest, ytrain, ytest
        """
    
        train = self.data[self.cols]
        target = self.data[self.name]
    
        return train_test_split(train, target, random_state=42, test_size=0.2, shuffle=True)
    
    def learner_selection(self):
        """This function calculates different score measures such as cross-validation, 
        determination coefficient (R2), root mean squared error (RMSE), and mean absolute error (MAE).
        listof_model : dictionary containing different model algorithms.
        It returns the calculated scores for each model.
        """ 
    
        result = {}
        
        x, _, y, _ = self.split_data() # Take only xtrain and ytrain.
    
        for cm in list(self.listof_model.items()):
        
            name = cm[0]
            model = cm[1]
        
            cvs = cross_val_score(model, x, y, cv=10).mean()
            ypred = cross_val_predict(model, x, y, cv=10)
            r2 = r2_score(y, ypred)
            mse = mean_squared_error(y, ypred)
            mae = mean_absolute_error(y, ypred)
            rmse = np.sqrt(mse)
        
            result[name] = {'cross_val_score': cvs, 'rmse': rmse, 'mae': mae, 'r2': r2}
        
            print('{} model done !!!'.format(name))
        
        return pd.DataFrame(result)
    
    def training_evaluate(self, algorithm):
        """This function trains and evaluates our model to find the determination coefficient (R2),
        root mean squared error (RMSE), and mean absolute error (MAE)."""
        
        result = {}
        xtrain, xtest, ytrain, ytest = self.split_data()
        
        learner = algorithm # learner = selected algorithm in the model_selection function
        
        model = learner.fit(xtrain, ytrain)
        ypred = model.predict(xtest)
        
        r2 = learner.score(xtest, ytest)
        rmse = np.sqrt(mean_squared_error(ytest, ypred))
        mae = mean_absolute_error(ytest, ypred)
        
        result['potential'] = {'r2': round(r2, 3), 'rmse': round(rmse, 3), 'mae': round(mae, 3)}
        
        return pd.DataFrame(result)
    
      
```

```python

def top_team(club=None, player_df_rd = None, n = 6):
    """
    This function gives the top team that are similar.
    """
    def cosine_similarity(a, b): 
        return a.dot(b.T) / (np.linalg.norm(a, 2) * np.linalg.norm(b, 2))
    
    df = {}
    
    data = player_df_rd.groupby('club')[cols].agg('mean').reset_index()
    
    pyers = data[data.club == club][cols].values[0]
    
    all_teams = list(set(data.club.values) - set([club])) 
    
    for u in all_teams:
        xv = cosine_similarity(pyers, data[data.club == u][cols].values[0])
        tn = data[data.club == u].club.values[0]
        df[u] = {club: round(xv, 3), 'club': tn}
        
    xd = pd.DataFrame(df).sort_values(by=club, axis=1, ascending=False)

    return xd.T[:n]
```

# Evaluation of Regression Models for Player Potential

An overview of evaluation metrics for different regression models applied to predicting player potential based on attributes such as age, overall score, and number of hits.

## Evaluated Models

- **LinearRegression**: Basic linear regression model.
- **KNeighborsRegression**: K-nearest neighbors regression, a non-parametric method.
- **RandomForestRegression**: Regression model based on random forests, an ensemble learning method.
- **GradientBoostingRegression**: Regression model based on gradient boosting, another ensemble learning technique.
- **XGBoostRegression**: XGBoost, an optimized library for gradient boosting.
- **Adaboost**: AdaBoost, another ensemble learning technique.

## Evaluation Metrics

The following metrics were used to evaluate the performance of each regression model:

- **cross_val_score**: Cross-validated score using techniques like k-fold cross-validation.
- **RMSE (Root Mean Squared Error)**: Measure of the average of the squares of errors or deviations.
- **MAE (Mean Absolute Error)**: Measure of the average of absolute errors between predicted and actual values.
- **R2 (R-squared)**: Indicates the proportion of the variance in the dependent variable that is predictable from the independent variables.

## Summary of Results

- **RandomForestRegression**, **GradientBoostingRegression**, and **XGBoostRegression** show overall good performances across all metrics, with **GradientBoostingRegression** and **XGBoostRegression** slightly outperforming the others.
- **KNeighborsRegression** also demonstrates good performances but not as consistent as ensemble methods.
- **LinearRegression** and **Adaboost** have relatively poor performances compared to other models, especially in terms of RMSE and MAE.

## Recommendations

- If you're looking for a model balancing performances across various metrics, consider **GradientBoostingRegression** or **XGBoostRegression**.

# Code Objective

The objective of this code is to group football players into homogeneous teams using the K-means algorithm. Players are characterized by their age, overall score, and number of hits. The ultimate goal is to form balanced teams in terms of positions and overall player performances.

## Features

- **calculate_inertia**: Calculates the inertia for a given number of clusters.
- **plot_elbow**: Plots the elbow graph to help determine the optimal number of clusters.
- **plot_clusters**: Plots scatter diagrams to visualize the distribution of players in the clusters.
- **regroup_players_by_position**: Groups players in each cluster by position.
- **form_teams_based_on_position**: Forms teams by selecting the best players from each position.





```python
# Method to calculate the inertia of the K-means model for a given number of clusters
def calculate_inertia(X, k):
    kmeans = KMeans(n_clusters=k, random_state=0, n_init='auto')
    kmeans.fit(X)
    return kmeans.inertia_

# Method to plot the elbow graph to help determine the optimal number of clusters
def plot_elbow(X, max_k):
    inertias = []
    for k in range(1, max_k + 1):
        inertias.append(calculate_inertia(X, k))
    plt.figure(figsize=(10, 6))
    plt.plot(range(1, max_k + 1), inertias, marker='o')
    plt.xlabel('Number of clusters')
    plt.ylabel('Inertia')
    plt.title('Elbow Method')
    plt.show()

# Method to plot cluster scatter graphs
def plot_clusters(X, max_k):
    fig, axes = plt.subplots(max_k // 2, 2, figsize=(15, 20))
    axes = axes.flatten()
    for k in range(2, max_k + 1):
        ax = axes[k-2]
        kmeans = KMeans(n_clusters=k, random_state=0, n_init='auto')
        kmeans.fit(X)
        cluster_labels = kmeans.labels_
        X_with_clusters = X.copy()
        X_with_clusters['Cluster'] = cluster_labels
        sns.scatterplot(ax=ax, data=X_with_clusters, x='position_encoded', y='overall', hue='Cluster', palette='viridis', legend=False)
        ax.set_title(f'k = {k}')
        ax.set_xlabel('Position')
        ax.set_ylabel('Overall')
        ax.set_xticks(range(len(position_encoder.classes_)))
        ax.set_xticklabels(position_encoder.classes_)
        
        # Plot red circles at the center of each cluster
        cluster_centers = kmeans.cluster_centers_
        for center in cluster_centers:
            ax.scatter(center[0], center[1], s=200, c='red', marker='o', edgecolors='black', linewidths=1, alpha=0.8)
        
        # Annotate the graph with cluster numbers
        for cluster_number in range(k):
            cluster_center = kmeans.cluster_centers_[cluster_number]
            ax.text(cluster_center[0], cluster_center[1], str(cluster_number+1), fontsize=12, color='black', ha='center', va='center')
            
    plt.tight_layout()
    plt.show()

# Method to form teams with players closest to the cluster centers
def form_teams(X, max_k, df):
    teams = []
    for k in range(2, max_k + 1):
        kmeans = KMeans(n_clusters=k, random_state=0, n_init='auto')
        kmeans.fit(X)
        cluster_labels = kmeans.labels_
        df_with_clusters = df.copy()
        df_with_clusters['Cluster'] = cluster_labels
        
        cluster_centers = kmeans.cluster_centers_
        for cluster_number in range(k):
            cluster_center = kmeans.cluster_centers_[cluster_number]
            # Identify players closest to the cluster center
            players_in_cluster = df_with_clusters[df_with_clusters['Cluster'] == cluster_number].copy()  
            distances_players = ((players_in_cluster[['position_encoded', 'overall']] - cluster_center) ** 2).sum(axis=1)
            closest_player = players_in_cluster.iloc[distances_players.idxmin()]
            # Add players to the team
            team = {
                'Cluster': cluster_number + 1,
                'Players': closest_player['name'],
                'Position': position_encoder.inverse_transform([int(closest_player['position_encoded'])])[0],
                'Overall': closest_player['overall']
            }
            teams.append(team)
    return pd.DataFrame(teams)

# Method to group players within each cluster by position and select the best player at each position
def group_players_by_position(X, max_k, df):
    grouped_players = defaultdict(list)
    for k in range(2, max_k + 1):
        kmeans = KMeans(n_clusters=k, random_state=0, n_init='auto')
        kmeans.fit(X)
        cluster_labels = kmeans.labels_
        df_with_clusters = df.copy()
        df_with_clusters['Cluster'] = cluster_labels
        
        for cluster_number in range(k):
            players_cluster = df_with_clusters[df_with_clusters['Cluster'] == cluster_number].copy()
            positions = players_cluster['position_encoded'].unique()
            for position in positions:
                players_at_position = players_cluster[players_cluster['position_encoded'] == position]
                best_player_at_position = players_at_position.loc[players_at_position['overall'].idxmax()]
                grouped_players[cluster_number].append(best_player_at_position)
    return grouped_players

# Method to form teams by selecting the best players of each position in each cluster
def form_teams_by_position(grouped_players):
    teams = []
    selected_positions = set()
    for cluster_number, players in grouped_players.items():
        for player in players:
            position = player['position_encoded']
            if position not in selected_positions:
                teams.append({
                    'Cluster': cluster_number + 1,
                    'Players': player['name'],
                    'Position': position_encoder.inverse_transform([int(position)])[0],
                    'Overall': player['overall']
                })
                selected_positions.add(position)
    return pd.DataFrame(teams)

# Encode position feature
position_encoder = preprocessing.LabelEncoder()
Player_df_rd['position_encoded'] = position_encoder.fit_transform(Player_df_rd['position'])

# Extract relevant features for clustering
X = Player_df_rd[['position_encoded', 'overall']]

# Plot the elbow visualization
max_k = 5
plot_elbow(X, max_k)

# Plot scatter plots for different k values
plot_clusters(X, max_k)

# Group players within each cluster based on position and overall rating
grouped_players = group_players_by_position(X, max_k, Player_df_rd)

# Form teams based on position and best overall rating
teams = form_teams_by_position(grouped_players)
print("Teams formed based on position and best overall rating:")
print(teams)


```
